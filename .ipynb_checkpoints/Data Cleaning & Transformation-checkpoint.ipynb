### Remove '\n' from the club column

```python
df['Club']
```

```python
# Remove "\n" from all rows in the 'Club' column
df['Club'] = df['Club'].apply(lambda x: x.strip())
```

```python
print(df['Club'])
```

### Remove '★' from the W/F, SM and IR column

```python
df['W/F'] = df['W/F'].apply(lambda x: x.replace("★", ""))
df['SM'] = df['SM'].apply(lambda x: x.replace("★", ""))
df['IR'] = df['IR'].apply(lambda x: x.replace("★", ""))
# Convert 'W/F','SM' and 'IR' column from object to int
df['W/F'] = df['W/F'].astype(int)
df['SM'] = df['SM'].astype(int)
df['IR'] = df['IR'].astype(int)
```

```python
print(df['W/F'])
print(df['SM'])
print(df['IR'])
```

```python
# Fix Value, Wage, Release Clause
def fix_Currency(column, df):
    Temp = []
    for x in range(len(df[column])):
        C = df[column][x]
        C = C.replace('€','')
        C = C.replace('K','000')
        C = C.replace('M','000000')
        C = C.replace('.','F')
        if  'F' in C:
            C = C.replace('F','')
            C = C = int(C) / 10
        Temp.append(int(C))
    
    df[column]= Temp
    print(df[column])
```

```python
fix_Currency('Value', df)
fix_Currency('Wage', df)
fix_Currency('Release Clause', df)
```

### Clean the Hits Column up

```python
#cleaning hits column from object format to int objects 
hits = []
for x in range(len(df["Hits"])):
    hit_value = df["Hits"][x]
    
    # Skip processing if the value is NaN
    if pd.isna(hit_value):
        hits.append(0)  # Or append 0 or another placeholder if desired
        continue
        
    # Replace 'K' with '000' and remove any '.' characters
    hit_value = hit_value.replace('K','000')
    hit_value = hit_value.replace('.','')
    
    # Convert the processed string back to an integer
    hits.append(hit_value)

df['Hits'] = hits
df['Hits'] = df['Hits'].astype(int)
print(df['Hits'])
```

### Make Height & Weight column numerical
Some values are in feets and inches and need to be converted to CM

```python
# Fix Height and Weight
def convert_height_to_cm(height):
    
    try:
        # Split the string into feet and inches
        feet, inches = height.split("'")
        inches = inches.replace('"', '')  # Remove the double quote

        # Convert feet to inches and then to cm
        total_inches = int(feet) * 12 + int(inches)
        total_cm = total_inches * 2.54  # 1 inch = 2.54 cm
        return round(total_cm, 0)  # Round to 2 decimal places
    except ValueError:
        return height  # Return None if the string format is not as expected

# Function to convert weight from lbs to kg
def convert_weight_to_kg(weight):

    try:
        # Remove the 'lbs' part and convert to float
        weight_in_lbs = float(weight.replace('lbs', ''))
        
        # Convert lbs to kg
        weight_in_kg = weight_in_lbs * 0.453592
        return round(weight_in_kg, 2)  # Round to 2 decimal places
    except ValueError:
        return weight  # Return None if the string format is not as expected
        
def fix_W_H(column, df):
    Temp = []
    for x in range(len(df[column])):
        C = df[column][x]
        if column == 'Height':
            C = C.replace('cm','')
            C = convert_height_to_cm(C)
        elif column == 'Weight':
            C = C.replace('kg','')
            C = convert_weight_to_kg(C)
            
        Temp.append(C)
    
    df[column]= Temp
    df[column] = df[column].astype(int)
    print(df[column])

fix_W_H('Height', df)
fix_W_H('Weight', df)
```

### Convert the joined column to Datetime

```python
# Convert to datetime using strptime
Temp_Joined = []
for x in range(len(df['Joined'])):
    Joined = df['Joined'][x]
    date = datetime.strptime(Joined, '%b %d, %Y')
    Temp_Joined.append(date)
    
df['Joined']= Temp_Joined
print(df['Joined'])
```
